__Outline:__ Derive list of known Transcription Factors in Mus musculus.

```python
import pandas as pd
```

The list of transcription factors is available from TFCat portal (http://www.tfcat.ca)

```python
tf_raw = pd.read_csv('export.txt', delimiter='\t', error_bad_lines=False, encoding='ISO-8859-1')
```

```python
tf_raw.head()
```

```python
tfs = tf_raw[['Gene ID', 'Evidence Strength']].drop_duplicates().dropna()
tfs['ID'] = list(map(int, tfs['Gene ID']))
```

These TFs are available as EntrezGene identifiers. They not to be converted to MGI symbols which can be done via: https://david.ncifcrf.gov/conversion.jsp .

```python
conv_tfs = pd.read_csv('conv_B1DE4DB27D351516106646392.txt', delimiter='\t')
```

```python
conv_tfs.head()
```

```python
def extract_symbol(name):
    s_idx = name.rfind('(')
    e_idx = name.rfind(')')
    return name[s_idx+1:e_idx]

conv_tfs['Gene Name'].apply(extract_symbol).to_csv('mm_tfs.txt', index=False)
```
